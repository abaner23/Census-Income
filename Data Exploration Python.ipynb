The adult dataset was first cited in a paper with title “Scaling Up the Accuracy of Naive-Bayes Classifiers: a Decision-Tree Hybrid”.  This dataset was first used to show that the NBTree classifier, which is a hybrid of C4.5 and Naïve-Bayes classifier, introduced in the above paper has a better accuracy rate than both C4.5 and Naïve-Bayes when used individually. 
This data was collected from the census bureau database of 1994. This dataset aims to predict whether a person’s income exceeds $50K/yr based on several attributes. We know that the income of a person will depend on several factors like education, type of job, location of job and others as well. This dataset gives information about incomes of people and the values of the attributes. The dataset will help to establish a relationship between yearly income and the useful attributes which could explain/predict the incomes of people for future purposes.

The dataset contains 14 attributes or independent variables and one dependent variable. Thus, this refers to a supervised learning problem. The attributes are described later. 

So each record contains the values of a person’s 14 attributes and the dependent variable value of whether he/she earns less or more than 50k/yr.  There are 32561 training data records in the dataset and 16281 test data records without removing the missing values.

To test our model, we can first find whether our model fits the data well. Depending upon the model used we can use a goodness-of-fit statistic to check the model fitting. For example, deviance can be used to check the model fit in case of logistic regression and R-squared statistic in case the model is linear regression. Once we are sure that our model fits the training data very well, we can evaluate the accuracy of our model by using test dataset and comparing its output with the actual output. If the accuracy is high then we are confident that we have been able to mine useful information from the data.

We can also get other types of information from the data. We can calculate proportions of people earning beyond 50k/year in each country and show a histogram for it. We might also be able to find useful patters within the attributes themselves for example between hours-per-week and country. This will give important information about the average working hours in each country.

Number of records with label=’>50k’= 7841(24% of the training dataset)
Number of records with label=’<=50k’= 24720(76% of the training dataset)
Since the number of records for label=’>50k’ are significantly large in number and the class probabilities reflect the overall income population distribution we can go ahead with this dataset for training the model.
There are 24 duplicate rows which can be kept in the training data set since looking at the data set two different people can have the exact same values of the attributes. Most probably it is not an error.
There are 4262 missing values out of which 1836 are from WorkClass, 1843 from occupation and 583 from native country.  Since there is no way to judge the country of origin and over 90% of the total dataset belong to the United States, there is an error probability of only 10% if we label all the missing native country column values as United States. We therefore enter United States for the 583 missing values.  In the same way, Private is the mostly occurring (69.7%) WorkClass value and thus we can impute the missing values with Private value. In case of occupation there is no clear majority value. Since intuitively the occupation plays a key role in deciding the income we cannot remove the attribute. Since we have over 32000 values, we can sacrifice the 1843 data rows. So, we remove the rows with missing values.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mp
import numpy as np
import scipy.stats as stat
import seaborn as sbn
df=pd.read_csv('E:\\Datasets\\adult_edited.csv',header=-1)
age=df.iloc[:,0]
agecat=pd.cut(age,bins=18)
df['agecat'] = agecat
dfg = df.groupby(by=14)
ds1=dfg['agecat'].value_counts()[0].sort_index()
x=np.arange(1,19,1)
ds2=dfg['agecat'].value_counts()[1].sort_index()
l1=np.array(ds1)
l2=np.array(ds2)
plt.figure(figsize=[10,10])
plt.bar(x,l1,color='gray')
plt.bar(x,l2,color='red',bottom=l1)
plt.xticks(np.arange(1,19,1))
plt.title('Class Count vs Age Categories')
gpatch=mp.Patch(color='gray',label='<=50k')


rpatch=mp.Patch(color='red',label='>50k')
plt.legend(handles=[gpatch,rpatch],loc='upper left')
plt.show()
```

```python
# Given a number (of hours) return a string (hours "bucket")
# based on whether it is less than, equal to, or greater than 40.
def tocat(h):
    return('<40' if h<40 else ('40' if h==40 else '>40') )

# Run the function above on every value in the hours-per-week column,
# then store the results in a new column called hrs_bkt
hrsweek=df.iloc[:,12]
df['hrs_bkt'] = hrsweek.apply(tocat)
x=np.sort(df['hrs_bkt'].unique())
dfg = df.groupby(by=14)
ds1=dfg['hrs_bkt'].value_counts()[0].sort_index()
ds2=dfg['hrs_bkt'].value_counts()[1].sort_index()
l1=np.array(ds1)
l2=np.array(ds2)
plt.figure(figsize=[10,10])
plt.bar(x,l1,color='gray')
plt.bar(x,l2,color='red',bottom=l1)
#plt.xticks(np.arange(1,19,1))
plt.title('Class Count vs Work hours per week')

gpatch=mp.Patch(color='gray',label='<=50k')


rpatch=mp.Patch(color='red',label='>50k')
plt.legend(handles=[gpatch,rpatch],loc='upper left')
plt.show()
```

The two new attributes that was created during the data exploration are:
    
1. WorkingHours_Categorical- The Working hours attribute was a continuous variable. After looking at the distribution of count of each class for all the values, it was clear that most people belonged to 40 hrs category. Also it was clear that proportion of people with lesser working hours and >50k income was low while opposite was true for more working hours. So I decided to change this attribute to a categorical variable with 3 categories: <40,40,>40.
2. Age_Categorical- The Age attribute was a continuous attribute but since income does not change significantly in 1-2 years the age atttribute was divided into bins. Since the age values ranged from 16 to 90, 18 bins was chosen with a width of 5 years.
